# Demonstração da Ferramenta

Fonte: https://www.kaggle.com/datasets/anshika2301/hr-analytics-dataset?resource=download

```python
# importações
import pandas as pd
```

```python
# leitura da base de dados
df = pd.read_csv('https://raw.githubusercontent.com/mrafaelbatista/masterclasse_alphatlytics/refs/heads/main/hr_analytics.csv')
```

```python
# Tradução das colunas em inglês
df.rename(columns={
    'EmpID': 'ID_Empregado',
    'Age': 'Idade',
    'AgeGroup': 'Faixa_Etaria',
    'Attrition': 'Atrito',
    'BusinessTravel': 'Viagem_a_Negocios',
    'DailyRate': 'Taxa_Diaria',
    'Department': 'Departamento',
    'DistanceFromHome': 'Distancia_de_Casa',
    'Education': 'Educacao',
    'EducationField': 'Campo_Educacional',
    'EmployeeCount': 'Contagem_Empregados',
    'EmployeeNumber': 'Numero_Empregado',
    'EnvironmentSatisfaction': 'Satisfacao_Ambiente',
    'Gender': 'Genero',
    'HourlyRate': 'Taxa_Por_Hora',
    'JobInvolvement': 'Envolvimento_Trabalho',
    'JobLevel': 'Nivel_Cargo',
    'JobRole': 'Cargo',
    'JobSatisfaction': 'Satisfacao_Cargo',
    'MaritalStatus': 'Estado_Civil',
    'MonthlyIncome': 'Renda_Mensal',
    'SalarySlab': 'Faixa_Salarial',
    'MonthlyRate': 'Taxa_Mensal',
    'NumCompaniesWorked': 'Numero_Empresas_Trabalhadas',
    'Over18': 'Mais_de_18',
    'OverTime': 'Hora_Extra',
    'PercentSalaryHike': 'Percentual_Aumento_Salarial',
    'PerformanceRating': 'Avaliacao_Performance',
    'RelationshipSatisfaction': 'Satisfacao_Relacionamento',
    'StandardHours': 'Horas_Padrao',
    'StockOptionLevel': 'Nivel_Opcao_Acoes',
    'TotalWorkingYears': 'Total_Anos_Trabalhados',
    'TrainingTimesLastYear': 'Treinamentos_Ultimo_Ano',
    'WorkLifeBalance': 'Equilibrio_Vida_Trabalho',
    'YearsAtCompany': 'Anos_na_Empresa',
    'YearsInCurrentRole': 'Anos_no_Cargo_Atual',
    'YearsSinceLastPromotion': 'Anos_Desde_Ultima_Promocao',
    'YearsWithCurrManager': 'Anos_Com_Gerente_Atual'
}, inplace=True)

# Exibir os nomes das colunas traduzidas para verificar
print(df.columns)
```

```python
# Agrupamento
agrupamento_departamento = df.groupby('Departamento')['Renda_Mensal'].mean()
display(agrupamento_departamento)
```

```python
# Ordenação
ordenacao_departamento = df.sort_values(by='Renda_Mensal', ascending=False).groupby('Departamento')['Renda_Mensal'].mean()
display(ordenacao_departamento)
```

```python
# Seleção dos dados sem os nulos
# df[df["col"].notna()]
df[df.notna()]
```

```python
# Seleção de dados
df[df["Campo_Educacional"].isin(['Medical', 'Marketing'])].sort_values(by='Campo_Educacional', ascending=False)
```

```python
# loc é baseado em rótulos
df.loc[df["Idade"] > 30, ["ID_Empregado", "Departamento" ]]
```

```python
# iloc baseado em inteiros (colunas e linhas)
df.iloc[9:25, 2:5]
```
